```python
import numpy as np
import time
```

# 1. Vector

## 1.1. Vector creation

```python
a = np.zeros(4)
print(f"np.zeros(4): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.zeros((4, ))
print(f"np.zeros((4, )): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.random.rand(4)
print(f"np.random.rand(4): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.arange(4.)
print(f"np.arrange(4): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.random.random_sample(4)
print(f"np.random.random_sample(4): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.array([5, 4, 3, 2])
print(f"np.array([5, 4, 3, 2]): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.array([5., 4, 3, 2])
print(f"np.array([5., 4, 3, 2]): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
```

## 1.2. Operations on vector

### 1.2.1. Indexing

```python
a = np.arange(10)
print(f"np.arrange(10): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")

print(f"a[2].shape: {a[2].shape}, a[2] = {a[2]}, accessing an element returns a scalar")
print(f"a[-1] = {a[-1]}")

try:
    c = a[10]
except Exception as e:
    print(e)
```

### 1.2.2. Slicing

```python
a = np.arange(10)
print(f"a = {a}")

c = a[2:7:1]
print(f"a[2:7:1] = {c}")

c = a[2:7:2]
print(f"a[2:7:2] = {c}")

c = a[3:]
print(f"a[3:] = {c}")

c = a[:3]
print(f"a[:3] = {c}")

c = a[:]
print(f"a[:] = {c}")
```

### 1.2.3. Single vector operations

```python
a = np.array([1, 2, 3, 4])
print(f"a = {a}")

b = -a
print(f"-a = {b}")

b = np.sum(a)
print(f"np.sum(a) = {b}")

b = np.mean(a)
print(f"np.mean(a) = {b}")

b = a**2
print(f"a**2 = {b}")
```

### 1.2.4. Vector element-wise operations

```python
a = np.array([1, 2, 3, 4])
b = np.array([-1, 2, -5, 7])
print(f"Binary operations work element wise: {a + b}")

c = np.array([1, 0])
try:
    d = a + c 
except Exception as e:
    print(e)
```

### 1.2.5. Scalar Vector Operations

```python
a = np.array([1, 2, 3, 4])
b = 5 * a
print(f"5 * a = {b}")
```

### 1.2.6. Vector dot product

```python
a = np.array([1, 2, 3, 4])
b = np.array([-1, 4, 3, 2])
c = np.dot(a, b)
print(f"NumPy 1-D np.dot(a, b) = {c}, np.dot(a, b).shape = {c.shape} ") 
c = np.dot(b, a)
print(f"NumPy 1-D np.dot(b, a) = {c}, np.dot(a, b).shape = {c.shape} ")
```

```python
def my_dot(a, b): 
    x=0
    for i in range(a.shape[0]):
        x = x + a[i] * b[i]
    return x

a = np.array([1, 2, 3, 4])
b = np.array([-1, 4, 3, 2])
print(f"my_dot(a, b) = {my_dot(a, b)}")
```

### 1.2.7. Vector vs for loop

```python
np.random.seed(1)
a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a, b)
toc = time.time()
print(f"np.dot(a, b) = {c:.4f}")
print(f"Vectorized version duration: {1000*(toc - tic):.4f}ms")


tic = time.time()
c = my_dot(a, b)
toc = time.time()
print(f"my_dot(a, b) = {c:.4f}")
print(f"Loop version duration: {1000*(toc - tic):.4f}ms")

del(a)
del(b)
```

# 2. Matrix

## 2.1. Matrix creation

```python
a = np.zeros((1, 5))                                       
print(f"a shape = {a.shape}, a = {a}")                     

a = np.zeros((2, 1))                                                                   
print(f"a shape = {a.shape}, a = {a}") 

a = np.random.random_sample((1, 1))  
print(f"a shape = {a.shape}, a = {a}") 
```

## 2.2. Matrix operation

### 2.2.1. Indexing

```python
a = np.arange(6).reshape(-1, 2)   # = np.arange(6).reshape(3, 2)
print(f"a.shape: {a.shape}, \na= \n{a}")

print(f"\na[2,0].shape:   {a[2, 0].shape}, a[2,0] = {a[2, 0]},     type(a[2,0]) = {type(a[2, 0])} Accessing an element returns a scalar\n")

#access a row
print(f"a[2].shape:   {a[2].shape}, a[2]   = {a[2]}, type(a[2])   = {type(a[2])}")
```

### 2.2.2. Slicing

```python
a = np.arange(20).reshape(-1, 10)
print(f"a = \n{a}")

print("\na[0, 2:7:1] = ", a[0, 2:7:1], ",  a[0, 2:7:1].shape =", a[0, 2:7:1].shape, "a 1-D array")

print("\na[:, 2:7:1] = \n", a[:, 2:7:1], ",  a[:, 2:7:1].shape =", a[:, 2:7:1].shape, "a 2-D array")

print("\na[:,:] = \n", a[:,:], ",  a[:,:].shape =", a[:,:].shape)

# access all elements in one row (very common usage)
print("\na[1,:] = ", a[1,:], ",  a[1,:].shape =", a[1,:].shape, "a 1-D array")
# same as
print("\na[1]   = ", a[1],   ",  a[1].shape   =", a[1].shape, "a 1-D array")
```
